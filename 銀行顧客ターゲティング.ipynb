# ラベルエンコーディング

```python
train_df = train.copy()
test_df = test.copy()
```

```python
train.head()
```

```python
test.head()
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

#ラベルエンコーディングの初期化とトレーニングデータのへの適応
label_encoder = LabelEncoder()
for colum in train_df.select_dtypes(include=['object']).columns:
  #y列がテストデータにないためエンコードから除外
  if colum != 'y':
    #トレーニングデータに基づいてフィット
    train_df[colum] = label_encoder.fit_transform(train_df[colum])
    #同じエンコーダーを使ってテストデータを変換
    test_df[colum] = label_encoder.transform(test_df[colum])

#返還後のトレーニングデータとテストデータを表示
display(train_df)
display(test_df)
```

# 学習モデルの構築
LightGBM


```python
import lightgbm as lgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

#訓練データとテストデータを特徴量とラベルに分類
x_train = train_df.drop(['y'], axis=1)
y_train = train_df['y']

# LightGBMが扱えるデータセット形式に変換
train_data = lgb.Dataset(x_train, label=y_train)

# パラメータの設定
params = {
    'objective': 'binary',
    'metric': 'binary_logloss',
    'boosting': 'gbdt',
    'num_leaves': 31,
    'learning_rate': 0.05,
    'verbose': 10
}

# モデルの訓練
gbm = lgb.train(params,
                train_data,
                num_boost_round=100)
```

```python
test_df.head()
```

# 予測

```python
# テストデータの特徴量
x_test = test_df

# テストデータでの予測（ここでは、y_testは実際には利用できないため、予測値の生成のみを示します）
y_pred = gbm.predict(x_test, num_iteration=gbm.best_iteration)

# 予測値を確率から二値（0または1）に変換
y_pred_binary = [1 if x > 0.5 else 0 for x in y_pred]
```

```python
y_pred_binary
```

```python
sub[1] = y_pred_binary
sub.head()
```
